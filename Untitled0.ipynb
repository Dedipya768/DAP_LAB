```python
def sum_of_digits(n):
    n = abs(n)   # handle negative numbers
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s

# Example
num = int(input("Enter an integer: "))
print("Sum of digits:", sum_of_digits(num))
```

```python
def product_of_digits(n):
    n = abs(n)   # handle negative numbers
    product = 1

    if n == 0:
        return 0

    while n > 0:
        product *= n % 10
        n //= 10

    return product

# Example
num = int(input("Enter an integer: "))
print("Product of digits:", product_of_digits(num))
```

```python
def reverse_number(n):
    sign = -1 if n < 0 else 1   # handle negative numbers
    n = abs(n)

    rev = 0
    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10

    return sign * rev

# Example
num = int(input("Enter a number: "))
print("Reversed number:", reverse_number(num))
```

```python
def is_palindrome(n):
    original = n
    n = abs(n)   # handle negative numbers

    rev = 0
    while n > 0:
        rev = rev * 10 + (n % 10)
        n //= 10

    return original == rev

# Example
num = int(input("Enter a number: "))
if is_palindrome(num):
    print("Palindrome number")
else:
    print("Not a palindrome number")
```
